### OUTLIERS

Neste caso, OUTLIERS acima fazem sentido \
Porém, há itens inválidos e muito abaixo do preço

Usar interquantil neste caso não é o método mais eficiente

```python
# numpy

df['preco'] = df['preco'].str.replace('R$ ', '').str.replace('.', '').str.replace(',', '.')
df['preco'] = df['preco'].astype(float)

df.info()

q1 = df['preco'].quantile(0.25)
q3 = df['preco'].quantile(0.75)

iqr = q3 - q1

limite_minimo = q1 - 1.5 * iqr
limite_maximo = q3 + 1.5 * iqr

acima_outliers = df[df['preco'] > limite_maximo]
abaixo_outliers = df[df['preco'] < limite_minimo]

display(acima_outliers, abaixo_outliers)

print(limite_maximo, limite_minimo)
print(df['preco'].mean())
```

### Separação por marca e modelo

```python
df['nome'] = df['nome'].str.replace('placa de vídeo ', '', case=False).str.replace('placa de video ', '', case=False)
df
```

```python
marcas = [
    'ASUS', 'MSI', 'Gigabyte', 'EVGA', 'Zotac', 'Palit', 'Galax', 'Sapphire', 'XFX', 'PowerColor', 'PNY', 'Leadtek', 'Inno3D', 'KFA2', 'Biostar', 'HIS', 'Club3D', 'VTX3D', 'ASRock', 'Shenzhen Zotac', 'Galaxy', 'Maxsun', 'Albatron', 'PCYes', 'Vx Pro', 'Afox', 'Aitek', 'Duex'
]

def extrair_marca(texto):
    texto = texto.lower()
    for marca in marcas:
        if marca.lower() in texto:
            return marca
    return 'Desconhecida'

df['marca'] = df['nome'].apply(extrair_marca)

df
```

```python
fabricantes_chip = [
    'NVIDIA', 'AMD', 'Intel'
]

def extrair_chip(texto):
    texto = texto.lower()
    for chip in fabricantes_chip:
        if chip.lower() in texto:
            return chip
        elif 'RX'.lower() in texto:
            return 'AMD'
        elif 'GeForce'.lower() in texto or 'RTX' in texto or 'GTX' in texto or 'GT' in texto:
            return 'NVIDIA'
    return 'Desconhecido'

df['chip'] = df['nome'].apply(extrair_chip)
```

```python
# tripla segurança na detecção de maisculas e minusculas, descobrindo methods diferentes
df['memoria_gb'] = df['nome'].str.extract(r'(?i)(\d+)\s*(MB|mb|GB|gb)', flags=re.IGNORECASE)[0]

display(df[df['memoria_gb'].isna()])

df = df.dropna(subset=['memoria_gb'])
df['memoria_gb'] = df['memoria_gb'].astype(int)

# placas provavelmente escritas em MB
display(df[df['memoria_gb'] > 128])

df.loc[df['memoria_gb'] > 128, 'memoria_gb'] = df.loc[df['memoria_gb'] > 128, 'memoria_gb'] / 1024

df
```

```python
# RAY TRACING

placas_com_ray_tracing = ['ray tracing', 'rtx', 'rx', 'RX']

def ray_tracing(texto):
    texto = texto.lower()
    for placa in placas_com_ray_tracing:
        if placa.lower() in texto:
            return True
    return False

df['ray_tracing'] = df['nome'].apply(ray_tracing)
df
```
