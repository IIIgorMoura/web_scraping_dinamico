```python
# %pip install pandas
import pandas as pd
df = pd.read_excel('cadeiras.xlsx')

df
```

```python
# 57 cadeiras sem cor especificada é um dos objetivos
# pegar cores primarias e secundarias
# trans preço em numero float padrão US

# juntar tds cores em array
cores = ['preto', 'azul']

def extrair_cor(texto):
    texto = texto.lower()
    for cor in cores:
        if cor in texto:
            return cor
    return 'Desconhecida'

df['cor_principal'] = df['marca'].apply(extrair_cor)
df[['marca', 'cor_principal']].head(10)

display(df.info())

df['preco'] = df['preco'].str.replace('R$', '').str.strip().str.replace(',', '.')

df
```
