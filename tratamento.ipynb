### OUTLIERS

Neste caso, OUTLIERS acima fazem sentido \
Porém, há itens inválidos e muito abaixo do preço

Usar interquantil neste caso não é o método mais eficiente

```python
import numpy as np

df['preco'] = df['preco'].str.replace('R$ ', '').str.replace('.', '').str.replace(',', '.')
df['preco'] = df['preco'].astype(float)

df.info()

q1 = df['preco'].quantile(0.25)
q3 = df['preco'].quantile(0.75)

iqr = q3 - q1

limite_minimo = q1 - 1.5 * iqr
limite_maximo = q3 + 1.5 * iqr

acima_outliers = df[df['preco'] > limite_maximo]
abaixo_outliers = df[df['preco'] < limite_minimo]

display(acima_outliers, abaixo_outliers)

print(limite_maximo, limite_minimo)
print(df['preco'].mean())
```
